```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from statsmodels.api import OLS
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVC
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
columns_to_drop = [
    'Employee_Name', 'EmpID', 'DOB', 'DateofHire', 'DateofTermination', 'LastPerformanceReview_Date',
    'Position', 'State', 'Zip', 'Sex', 'MaritalDesc', 'CitizenDesc', 'HispanicLatino', 'RaceDesc',
    'TermReason', 'ManagerName', 'ManagerID', 'RecruitmentSource', 'PerformanceScore', 'Salary'
]

X = df.drop(columns = columns_to_drop)
y = df['Salary']
```

```python
X = pd.get_dummies(X,dtype=int, drop_first=True)
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
X_pca
```

```python
X_train_pca , X_test_pca , y_train , y_test = train_test_split(X_pca,y,test_size=0.2,random_state=30)
```

```python
ridge = Ridge(alpha=0.1)
ridge.fit(X_train_pca,y_train)

ridge_y_pred = ridge.predict(X_test_pca)

ridge_mse = mean_squared_error(y_test, ridge_y_pred)
ridge_mae = mean_absolute_error(y_test, ridge_y_pred)
ridge_rmse = np.sqrt(ridge_mse)
ridge_r2 = r2_score(y_test, ridge_y_pred)

print('Ridge Regularization Performance')
print(f'Mean Squared Error: {ridge_mse}')
print(f'Mean Absolute Error: {ridge_mae}')
print(f'Root Mean Squared Error: {ridge_rmse}')
print(f'R-squared: {ridge_r2}')
```

```python
ridge.score(X_test_pca,y_test)
```

```python
linear = LinearRegression()
linear.fit(X_train_pca, y_train)

linear_y_pred = linear.predict(X_test_pca)

linear_mse = mean_squared_error(y_test, linear_y_pred)
linear_mae = mean_absolute_error(y_test, linear_y_pred)
linear_rmse = np.sqrt(linear_mse)
linear_r2 = r2_score(y_test, linear_y_pred)

print('Linear Regression Performance')
print(f'Mean Squared Error: {linear_mse}')
print(f'Mean Absolute Error: {linear_mae}')
print(f'Root Mean Squared Error: {linear_rmse}')
print(f'R-squared: {linear_r2}')
```

```python
linear.score(X_test_pca,linear_y_pred)
```

```python
lasso = Lasso(alpha=0.1)
lasso.fit(X_train_pca,y_train)

lasso_y_pred = lasso.predict(X_test_pca)

lasso_mse = mean_squared_error(y_test, lasso_y_pred)
lasso_mae = mean_absolute_error(y_test, lasso_y_pred)
lasso_rmse = np.sqrt(lasso_mse)
lasso_r2 = r2_score(y_test, lasso_y_pred)

print('Lasso Regularization Performance')
print(f'Mean Squared Error: {lasso_mse}')
print(f'Mean Absolute Error: {lasso_mae}')
print(f'Root Mean Squared Error: {lasso_rmse}')
print(f'R-squared: {lasso_r2}')
```

```python
lasso.score(X_test_pca,y_test)
```

```python
import statsmodels.api as sm
X_train_const = sm.add_constant(X_train_pca)
```

```python
ols = sm.OLS(y_train, X_train_const).fit()
ols.summary()
```

```python
newX_test = sm.add_constant(X_test_pca)

ols_y_pred = ols.predict(newX_test)
```

```python
ols_mae = mean_absolute_error(y_test,ols_y_pred)
ols_mse = mean_squared_error(y_test,ols_y_pred)
ols_rmse = np.sqrt(ols_mse)
ols_r2 = r2_score(y_test,ols_y_pred)

print('OLS Regression Performance')
print(f'Mean Squared Error: {ols_mse}')
print(f'Mean Absolute Error: {ols_mae}')
print(f'Root Mean Squared Error: {ols_rmse}')
print(f'R-squared: {ols_r2}')
```

```python
model_info={
    'Model':['Linear Regression','Ridge Regression','Lasso Regression','OLS Method'],
    'MAE':[linear_mae,ridge_mae,lasso_mae,lasso_mae],
    'MSE':[linear_mse,ridge_mse,lasso_mse,lasso_mse],
    'RMSE':[linear_rmse,ridge_rmse,lasso_rmse,ols_rmse],
    'R2':[linear_r2,ridge_r2,lasso_r2,ols_r2],  
}
performance = pd.DataFrame(model_info)
performance
```
